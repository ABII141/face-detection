# import libraries

```python
import cv2
import matplotlib.pyplot as plt
```

# for single face image

```python
face1 = cv2.imread(r"C:\Users\LENOVO\Desktop\verzeo\veronica.jpg", 0)  
fd = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def get_face(face):  
    face_img = face.copy()  
    fr = fd.detectMultiScale(face_img)  
    for (x,y,width,height) in fr:  
        cv2.rectangle(face_img, (x,y), (x + width, y+height), (255,255,77), 10)                
    return face_img

result = get_face(face1)
plt.imshow(result)
plt.show()
```
